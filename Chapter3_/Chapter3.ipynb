>  This relatively short chapter focuses on an idea that is always bouncing around our heads, but is rarely made explicit outside books devoted to statistics or Monte Carlo. The idea's simplicity can sometimes be so simple as to be confusing, so we will go through some examples and make it a transparent tool.

The second hammer: *The Law of Large Numbers*
------

Let $Z_i$ be samples from some probability distribution. According to *the Law of Large numbers*,  so long as $E[Z]$ is finite, the following holds,

$$\frac{1}{N} \sum_{i=0}^N Z_i \rightarrow E[ Z ],  \;\;\; N \rightarrow \infty$$


This may seem like a boring result, but it will be the most useful tool you use.

### Intution 

If the above Law is somewhat surprising, or not inutitive, it can be made more clear be considering  a small example. 

Consider a random variable $Z$ that can take on only two values, $c_1$ and $c_2$, with probabilities $P(Z=c_1) = 0.25$ and $P(Z = c_2 ) = 0.75 )$. Suppose we have a large number of samples of $Z$, calling a specific sample $Z_i$. We can approximate the expected value of $Z$ by averaging. By definition, $Z_i$ can only take on $c_1$ or $c_2$, hence we can partition the sum over these two values:

\begin{align*}
\frac{1}{N} \sum_{i=0}^N \;Z_i = 
& \frac{1}{N} \big(  \sum_{ Z_i = c_1}c_1 + \sum_{Z_i=c_2}c_2 \big) \\\\[5pt]
& = c_1 \sum_{ Z_i = c_1}\frac{1}{N} + c_2 \sum_{ Z_i = c_2}\frac{1}{N} \\\\[5pt]
& = c_1 \times \text{ (approximate frequency of $c_1$) } + c_2 \times \text{ (approximate frequency of $c_2$) } \\\\[5pt]
& \approx c_1 \times P(Z = c_1) + c_2 \times P(Z = c_2 ) \\\\[5pt]
& = E[Z]
\end{align*}


Equality holds in the limit, but we can get very closer and closer by using more and more samples. This Law hold for any distribution, minus some pathological examples that will rarely occur. 

____
### Example

Below is a diagram of the Law of Large numbers in action for three different sequences of Poisson random variables. 

```python
figsize( 10, 5 )
sample_size = 100000
expected_value = lambda_ = 4.5
poi = stats.poisson
N_samples = range(0,sample_size,100)

samples = poi.rvs( lambda_, size = sample_size ) 
partial_average = [ samples[:i].mean() for i in N_samples ]
plt.plot( N_samples, partial_average, lw=1.5,label="partial average of samples")

samples = poi.rvs( lambda_, size = sample_size ) 
partial_average = [ samples[:i].mean() for i in N_samples ]
plt.plot( N_samples, partial_average,  lw=1.5, label="partial average of samples")

samples = poi.rvs( lambda_, size = sample_size ) 
partial_average = [ samples[:i].mean() for i in N_samples ]
plt.plot( N_samples, partial_average, lw=1.5, label="partial average of samples")


plt.plot( N_samples, expected_value*np.ones_like( partial_average), \
    ls = "--", label = "true expected value", c = "k" )

plt.ylim( 4.35, 4.65) 
plt.title( "Convergence of the average of \n random variables to its expected value" )
plt.ylabel( "average of samples" )
plt.xlabel( "# of samples")
plt.legend()
```

Looking at the above plot, it is clear that when the sample size is small, there is greater variation in the average (compare how *jagged and jumpy* the average is initially, then *smooths* out). All three paths *approach* the value 4.5, but just flirt with it as $N$ gets large. Mathematicians and statistician have another name for *flirting*: convergence. 

Another very relevant question we can ask is *how quickly am I converging to the expected value?* Let's plot something new. For a specific $N$, let's do the above trials thousands of times and compute how far away we are from the true expected value, on average. But wait-- *compute on average*? This simply the law of large numbers again! For example, we are interested in, for a specific $N$:

$$\sqrt{ \;E\left[\;\; \left( \frac{1}{N}\sum_{i=0}^NZ_i  - 4.5 \;\right)^2 \;\;\right] \;\;}$$

(We take the square root so the dimensions of the above quantity and our samples are the same). As the above is an expected value, this can be approximated using the law of large numbers! Instead of averaging $Z_i$, we use caluculate thousands of times:

$$ \left( \;\frac{1}{N}\sum_{i=0}^NZ_i  - 4.5 \; \right)^2 $$

```python
figsize( 10, 3.5)
n_samples = 250
expected_value_n = [] 
x = np.arange( 0, 50000, 2000 )
for N in x:
    #for a specific N, we need to calulate the above expected value.
    samples = poi.rvs( lambda_, size = (N, n_samples) )
    averages = samples.mean(axis=0)
    expected_value_n.append( ((averages - expected_value)**2).mean() )

plt.xlabel( "$N$" )
plt.ylabel( "expected squared-distance from true value" )
plt.plot(x,  np.sqrt(expected_value_n), lw = 3, label="expected squared distance" )
plt.plot( x, np.sqrt(expected_value)/np.sqrt(x), ls = "--", label = r"$\frac{\sqrt{\lambda}}{\sqrt{N}}$" )
plt.legend()
```

It turns out we can measure the rate of convergence. Above I have plotted a second line, the function $\sqrt{\lambda}/\sqrt{N}$. In most cases, given a sequence of random variable distributed like $Z$, the rate of converge to $E[Z]$ of the Law of Large Numbers is 

$$ \frac{ \sqrt{ \; Var(Z) \; } }{\sqrt{N} }$$


This is useful to know: for a given large $N$, we know (on average) for far away we are from the estimate. On the other hand, in a Bayesian setting this can seem like a oxymoron: Bayesian analysis is OK with uncertainity so what's the *statistical* point of adding an extra precise digits? Though drawing samples can be so computationally cheap that *larger* $N$ is fine too. 

#### What is $Var(Z)$ though?

The variance is simply another expected value that can be approximated! Consider the following, once we have the expected value (using the Law of Large Numbers, denote it $\mu$), we can estimate the variance:

$$ \frac{1}{N}\sum_{i=0}^N \;(Z_i - \mu)^2 \rightarrow E[ \;( Z - \mu)^2 \;] = Var( Z )$$




## What does this all have to do with Bayesian statistics? 


In more analytical Bayesian inference, we would of been required to evaluate complicated expected values represented as multi-dimensional integrals. No longer. If we can sample from the disitribution, we simply need to evaluate averages. Much easier. Plots like the ones above show how fast you are converging.
